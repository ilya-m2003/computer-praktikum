```python
from tkinter import *
from tkinter import font
from PIL import Image, ImageTk

n = 6
def get_word():   
    global guess_word, unknown_word, label, entry, button, entry1, label0
    guess_word = entry.get().split()[0]
    unknown_word = entry.get().split()[1]
    word(len(guess_word))
    for x in range(len(unknown_word)):
        if unknown_word[x] != "_":
            letter(unknown_word[x], x)
    label.destroy()
    entry.destroy()
    button.destroy()
    tk.update()
    label0 = Label(tk, text='Осталось попыток: ' + str(n), font=font1)
    label0.place(x=25, y=25, width=350, height=30)
    label1 = Label(tk, text='Введите угадываемую букву"')
    label1.place(x=25, y=350, width=400, height=30)
    entry1 = Entry(tk)
    entry1.place(x=450, y=350, width=100, height=30)
    button1 = Button(tk, command=guess_letter)
    button1.place(x=575, y=350, width=100, height=30)
    
def guess_letter():
    global guess_word, unknown_word, entry1, label0
    new_word = ""
    letter1 = entry1.get()
    for x in range(len(guess_word)):
        if letter1 == guess_word[x] or letter1 == unknown_word[x] or guess_word[x] == unknown_word[x]:
            new_word += guess_word[x]#letter1
        else:
            new_word += "_"
    if letter1 not in guess_word:
        global n
        n -= 1
        print(new_word, guess_word)
        if n == 0:
            global canvas1
            label0.config(text='Ты проиграл!!!')
            tk.img =img= ImageTk.PhotoImage(Image.open("проиграл.jpg"))
            canvas1.create_image(0,0,anchor='nw',image=img)
            canvas1.update()
        else:
            label0.config(text='Осталось попыток: ' + str(n))
        tk.update()
    unknown_word = new_word
    for x in range(len(unknown_word)):
        if unknown_word[x] != "_":
            letter(unknown_word[x], x)
            #print(1, unknown_word[x], x)
    if new_word == guess_word:
        label0.config(text='Ты выиграл!!!')
        tk.img = img = ImageTk.PhotoImage(Image.open("выиграл.jpg"))
        canvas1.create_image(0,0,anchor='nw',image=img)
        tk.update()
        canvas1.update()
    
tk = Tk()
tk.title("Палач")
tk.geometry("700x500")
font1 = font.Font(family= "Arial", size=20, weight="bold")
canvas = Canvas(tk, bg="black")
canvas.place(x=25, y=400, width=650, height=80)
canvas1 = Canvas(tk)
canvas1.place(x=25, y=100, width=200, height=200)
label = Label(tk, text='Введите загадываемое слово в формате: "ПРАКТИКА П__К____"')
label.place(x=25, y=350, width=350, height=30)
entry = Entry(tk)
entry.place(x=400, y=350, width=150, height=30)
button = Button(tk, command=get_word)
button.place(x=575, y=350, width=100, height=30)

def word(n):
    global x0, y0
    width = 60 * n - 10
    x0 = 25 + int((600 - width) * 0.5)
    y0 = 15
    for x in range(n):
        canvas.create_rectangle(x0 + 60 * x, y0, x0 + 60 * x + 50, y0 + 50, fill="#FFF") 
        canvas.update()
        
def letter(x, p):
    canvas.create_rectangle(x0 + 60 * p, y0, x0 + 60 * p + 50, y0 + 50, fill="#FFF")
    canvas.create_text(x0 + 60 * p + 25, y0 + 25, text=x, font=font1)
    canvas.update()
    
tk.mainloop()
```
