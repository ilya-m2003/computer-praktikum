```python
from tkinter import *
from tkinter import messagebox
from PIL import Image,ImageTk
import time 
board = [[" ", " ", " "],
         [" ", " ", " "],
         [" ", " ", " "]]
f = 0

def find_triple():
    o_points = []
    x_points = []
    if board[0] == ["O", "O", "O"]:
        o_points.append((0, 0))
        o_points.append((0, 1))
        o_points.append((0, 2))
    elif board[1] == ["O", "O", "O"]:
        o_points.append((1, 0))
        o_points.append((1, 1))
        o_points.append((1, 2))
    elif board[2] == ["O", "O", "O"]:
        o_points.append((2, 0))
        o_points.append((2, 1))
        o_points.append((2, 2))
    elif [board[0][0], board[1][0], board[2][0]] == ["O", "O", "O"]:
        o_points.append((0, 0))
        o_points.append((1, 0))
        o_points.append((2, 0))
    elif [board[0][1], board[1][1], board[2][1]] == ["O", "O", "O"]:
        o_points.append((0, 1))
        o_points.append((1, 1))
        o_points.append((2, 1))
    elif [board[0][2], board[1][2], board[2][2]] == ["O", "O", "O"]:
        o_points.append((0, 2))
        o_points.append((1, 2))
        o_points.append((2, 2))
    elif [board[0][0], board[1][1], board[2][2]] == ["O", "O", "O"]:
        o_points.append((0, 0))
        o_points.append((1, 1))
        o_points.append((2, 2))
    elif [board[0][2], board[1][1], board[2][0]] == ["O", "O", "O"]:
        o_points.append((0, 2))
        o_points.append((1, 1))
        o_points.append((2, 0))
    if board[0] == ["X", "X", "X"]:
        x_points.append((0, 0))
        x_points.append((0, 1))
        x_points.append((0, 2))
    elif board[1] == ["X", "X", "X"]:
        x_points.append((1, 0))
        x_points.append((1, 1))
        x_points.append((1, 2))
    elif board[2] == ["X", "X", "X"]:
        x_points.append((2, 0))
        x_points.append((2, 1))
        x_points.append((2, 2))
    elif [board[0][0], board[1][0], board[2][0]] == ["X", "X", "X"]:
        x_points.append((0, 0))
        x_points.append((1, 0))
        x_points.append((2, 0))
    elif [board[0][1], board[1][1], board[2][1]] == ["X", "X", "X"]:
        x_points.append((0, 1))
        x_points.append((1, 1))
        x_points.append((2, 1))
    elif [board[0][2], board[1][2], board[2][2]] == ["X", "X", "X"]:
        x_points.append((0, 2))
        x_points.append((1, 2))
        x_points.append((2, 2))
    elif [board[0][0], board[1][1], board[2][2]] == ["X", "X", "X"]:
        x_points.append((0, 0))
        x_points.append((1, 1))
        x_points.append((2, 2))
    elif [board[0][2], board[1][1], board[2][0]] == ["X", "X", "X"]:
        x_points.append((0, 2))
        x_points.append((1, 1))
        x_points.append((2, 0))
    if len(o_points) + len(x_points) > 0:
        time.sleep(0.2)   
    for x in o_points:
        canvas.create_rectangle(100*x[1], 100*x[0], 100*x[1]+100, 100*x[0]+100, fill="#0F0", outline="#0C0")
        canvas.update()
    for x in x_points:
        canvas.create_rectangle(100*x[1], 100*x[0], 100*x[1]+100, 100*x[0]+100, fill="#F46", outline="#F46")
        canvas.update()
    
    if len(x_points) > 0:
        return "X"
    elif len(o_points) > 0:
        return "O"
    elif " " not in board[0] and " " not in board[1] and " " not in board[2]:
        return "F"
    else:
        return False
    
def fill_board(x, y):
    global board, f, play_again, canvas_win
    if board[y][x] == " ":
        if f == 0:
            board[y][x] = "O"
        else:
            board[y][x] = "X"
        f = 1 - f
    else:
        messagebox.showerror('Ошибка', 'Неверный ход')
    
    if find_triple() == "O":
        time.sleep(1)
        canvas_win = Canvas(tk)
        canvas_win.place(x=0, y=0, width=300, height=300)
        tk.img =img= ImageTk.PhotoImage(Image.open("нолики выиграли.jpg"))
        canvas_win.create_image(0,0,anchor='nw',image=img)
        tk.update()
        canvas_win.update()
        canvas_win.bind('<Button-1>', play_again)
    elif find_triple() == "X":
        time.sleep(1)
        canvas_win = Canvas(tk)
        canvas_win.place(x=0, y=0, width=300, height=300)
        tk.img =img= ImageTk.PhotoImage(Image.open("крестики выиграли.jpg"))
        canvas_win.create_image(0,0,anchor='nw',image=img)
        tk.update()
        canvas_win.update()
        canvas_win.bind('<Button-1>', play_again)
    elif find_triple() == "F":
        canvas_win = Canvas(tk)
        canvas_win.place(x=0, y=0, width=300, height=300)
        tk.img =img= ImageTk.PhotoImage(Image.open("ничья.jpg"))
        canvas_win.create_image(0,0,anchor='nw',image=img)
        tk.update()
        canvas_win.update()
        canvas_win.bind('<Button-1>', play_again)
        
        
    
def render():
    for y in range(3):
        for x in range(3):
            if board[y][x] == "O":
                #canvas.create_line(100 * x + 48, 100 * y + 13, 100 * x + 22, 100 * y + 91, width=5)
                #canvas.create_line(100 * x + 22, 100 * y + 91, 100 * x + 85, 100 * y + 43, width=5)
                #canvas.create_line(100 * x + 85, 100 * y + 43, 100 * x + 7, 100 * y + 43, width=5)
                #canvas.create_line(100 * x + 7, 100 * y + 43, 100 * x + 72, 100 * y + 91, width=5)
                #canvas.create_line(100 * x + 72, 100 * y + 91, 100 * x + 48, 100 * y + 13, width=5)
                
                canvas.create_oval(100 * x + 20, 100 * y + 20, 100 * x + 80, 100 * y + 80, outline="#0C0", width=5)
                
                #canvas.create_rectangle(100 * x + 10, 100 * y + 20, 100 * x + 90, 100 * y + 40, fill="#FFF", outline="#FFF")
                #canvas.create_rectangle(100 * x + 10, 100 * y + 40, 100 * x + 90, 100 * y + 60, fill="#00F", outline="#00F")
                #canvas.create_rectangle(100 * x + 10, 100 * y + 60, 100 * x + 90, 100 * y + 80, fill="#F00", outline="#F00")
                
            if board[y][x] == "X":
                canvas.create_line(100 * x + 20, 100 * y + 20, 100 * x + 80, 100 * y + 80, fill="#F04", width=5)
                canvas.create_line(100 * x + 80, 100 * y + 20, 100 * x + 20, 100 * y + 80, fill="#F04", width=5)
                
                #canvas.create_rectangle(100 * x + 21, 100 * y + 29, 100 * x + 73, 100 * y + 41, fill="#F04", outline="#F04")
                #canvas.create_rectangle(100 * x + 41, 100 * y + 7, 100 * x + 53, 100 * y + 93, fill="#F04", outline="#F04")
                #canvas.create_rectangle(100 * x + 10, 100 * y + 20, 100 * x + 90, 100 * y + 50, fill="#00F", outline="#00F")
                #canvas.create_rectangle(100 * x + 10, 100 * y + 50, 100 * x + 90, 100 * y + 80, fill="#FF0", outline="#FF0")
            
def click(event):
    x = event.x // 100
    y = event.y // 100
    fill_board(x, y)
    render()
    
def play_again(event):
    global board, f, canvas_win, canvas
    del canvas_win, canvas
    board = [[" ", " ", " "],
         [" ", " ", " "],
         [" ", " ", " "]]
    f = 0 
    canvas = Canvas(tk)
    canvas.place(x=0, y=0, width=300, height=300)
    canvas.create_line(0, 100, 300, 100)
    canvas.create_line(0, 200, 300, 200)
    canvas.create_line(100, 0, 100, 300)
    canvas.create_line(200, 0, 200, 300)
    canvas.bind('<Button-1>', click)
    
tk = Tk()
tk.title("Крестики-нолики")
tk.geometry("300x300")
canvas = Canvas(tk)
canvas.place(x=0, y=0, width=300, height=300)
canvas.create_line(0, 100, 300, 100)
canvas.create_line(0, 200, 300, 200)
canvas.create_line(100, 0, 100, 300)
canvas.create_line(200, 0, 200, 300)
canvas.bind('<Button-1>', click)
tk.mainloop()
```
